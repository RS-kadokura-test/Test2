## Googleドライブに保存されたノートの「保存」を行う
Googleドライブに保存されたノートに対し、下記はRSリモート環境では実行不可。 
> [ファイル]タブ ⇒ [保存]

更新したファイルを保存する場合、GitHub上で保存を行うこと。手順は下記の「代用手段」に示す。

◆代用手段
1. 「GitHubにコピーを保存」、もしくは「コピーをGitHubGistとして保存」を実施する。

◆補足
「GitHubにコピーを保存」と「コピーをGitHubGistとして保存」の違い

*   「GitHubにコピーを保存」では、GitHubのリポジトリ/ブランチを選択し、名前をつけてノートブックを保存する。リポジトリ/ブランチ内に同名ファイルがある場合は変更扱いとなる。

*   「GitHubGistとして保存」では、現在のノートブックで新たなGitHubGistを作成する。
> GitHubGistはコード断片の共有サービスです。
> 新たにGitHubGistを作成するとそのGitHubGistのリポジトリが生成されます。<br>
> 作成したGitHubGistはsecretとなり（※）検索できません。URLがわかればアクセスできますが、（URLはGitHub側で生成したハッシュ値）
>※（初期設定では）
>「GitHubGistとして保存」をクリックするたびに新たなGitHubGistが作成されます。
